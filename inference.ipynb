### Naive Bayes Image Classifier

```python
from NaiveBayesImageClassifier import naive_bayes_image_classifer
import numpy as np
import math
import os
```

```python
path = 'C:/Users/User/Study/Projects/Data/MNIST'
trian_image_file = open(os.path.join(path, 'train-images-idx3-ubyte'), 'rb')
train_label_file = open(os.path.join(path, 'train-labels-idx1-ubyte'), 'rb')
test_image_file = open(os.path.join(path, 't10k-images-idx3-ubyte'), 'rb')
test_label_file = open(os.path.join(path, 't10k-labels-idx1-ubyte'), 'rb')

NumOfClass = 10
NumOfTrain = 60000
NumOfTest = 10000
h, w = 28, 28
c = 1
b = 32

X_train = np.zeros((NumOfTrain, h * w, c), dtype='uint8')
y_train = np.zeros(NumOfTrain, dtype='uint8')
X_test = np.zeros((NumOfTest, h * w, c), dtype='uint8')
y_test = np.zeros(NumOfTest, dtype='uint8')

trian_image_file.read(16)
train_label_file.read(8)

for i in range(NumOfTrain):
    for j in range(h*w):
        for k in range(c):
            X_train[i, j, k] = int.from_bytes(trian_image_file.read(1), byteorder='big')
    y_train[i] = int.from_bytes(train_label_file.read(1), byteorder='big')

test_image_file.read(16)
test_label_file.read(8)
for i in range(NumOfTest):
    for j in range(h*w):
        for k in range(c):
            X_test[i, j, k] = int.from_bytes(test_image_file.read(1), byteorder='big')
    y_test[i] = int.from_bytes(test_label_file.read(1), byteorder='big')
```

```python
X_train.shape
```

```python
Naive_classifier = naive_bayes_image_classifer(10)
```

```python
Naive_classifier.forward(X_train, y_train, X_test, y_test, Mode='continuous')
```
